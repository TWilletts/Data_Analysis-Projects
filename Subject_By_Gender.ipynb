```python
Female_list = []
Male_list = []
Ratio = []
Percent = []
Dominated = []
for i in range(24):
    Ratio.append(data.Number[i*2]/data.Number[(i*2)+1])
    Ratio.append(data.Number[i*2]/data.Number[(i*2)+1])
    Percent.append(data.Number[i*2]/(data.Number[i*2] + data.Number[(i*2)+1]))
    Percent.append(data.Number[i*2]/(data.Number[i*2] + data.Number[(i*2)+1]))
    if data.Number[i*2]/data.Number[(i*2)+1] > 1:
        Dominated.append('Female')
        Dominated.append('Female')
        Female_list.append(data['CAH level 1'][i*2])
    else:
        Dominated.append('Male')
        Dominated.append('Male')
        Male_list.append(data['CAH level 1'][i*2])
        
```

```python
data['Ratio'] = Ratio
```

```python
data['Percent'] = Percent
```

```python
data['Dominated by'] = Dominated
```

```python
for i in range(len(data)):
    data['Percent'][i] = round((data['Percent'][i]*100),2)
```

```python
data = data.sort_values(by=['Percent'], ascending=False)
data
```

```python
female_sum = 0
male_sum = 0

for i in range(len(data)):
    if data['Sex'][i] == 'Female':
        female_sum += data['Number'][i]
    if data['Sex'][i] == 'Male':
        male_sum += data['Number'][i]

print(female_sum, male_sum)
```
